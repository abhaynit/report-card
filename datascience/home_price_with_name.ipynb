```python
import pandas as pd
df = pd.read_csv("file\\patna.csv")
df.head()
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies],axis = 'columns')
merged
```

```python
final = merged.drop(['town','delhi'],axis = 'columns')
final
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(final[['area','kolkata','patna']],final.price)
```

```python
model.score(final[['area','kolkata','patna']],final.price)
```

```python
model.predict([[2600,1,0]])
```

```python
import pickle
with open("home_price_prediction_with_name_area",'wb') as f:
    pickle.dump(model,f)
```

```python
with open("home_price_prediction_with_name_area","rb") as f:
    ab = pickle.load(f)
```
